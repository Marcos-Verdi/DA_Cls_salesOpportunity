```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Spinach Price - Sheet1.csv',header=1)
df
```

```python
from datetime import date
df['date'] = pd.to_datetime(df.Fecha, format='%d/%m/%Y')
```

```python
df['Year'] = df['date'].dt.year
```

```python
df['Month'] = df['date'].dt.month
```

```python
df['Day'] = df['date'].dt.day
```

```python
df['Spinach Price'] = df['Espinaca']
```

```python
df = df.drop(['Fecha','Espinaca','date'],axis=1)
df
```

```python
for i in df.Year.unique():
    plt.figure(figsize=(20,5))
    by_year = df.loc[df.Year == i]
    sns.lineplot(x='Month',y='Spinach Price', data=by_year)
    plt.xticks(range(1,13))
    plt.title(i)
```

```python
df2 = pd.read_csv('Spinach_Price_Quantity - Sheet1.csv', header=0)
df2
```

```python
df2['Date'] = pd.to_datetime(df2.Date, format='%d/%m/%Y')
```

```python
df2['Year'] = df2['Date'].dt.year
```

```python
df2['Month'] = df2['Date'].dt.month
```

```python
df2['Day'] = df2['Date'].dt.day
```

```python
df2 = df2.drop('Date',axis=1)
df2
```

```python
for i in df2.Year.unique():
    plt.figure(figsize=(20,5))
    by_year = df2.loc[df2.Year == i]
    sns.scatterplot(x='Volume',y='Spinach Price', data=by_year)
    plt.title(i)
```
